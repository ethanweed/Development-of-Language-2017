```python
import glob
from os import chdir as cd
child = 'Adam'
pathin = '/Users/ethan/Desktop/Brown/'+ child
funcpath = '/Users/ethan/Documents/Scripts/Teaching/Development-of-Language/DoL_2017/Functions'

cd(funcpath)
import MLU

cd(pathin)
```

```python
kid = []
mom = []
for file in glob.glob('*.cha'):
    c_mlu = MLU.CHI(file)
    m_mlu = MLU.MOT(file)
    kid.append(c_mlu)
    mom.append(m_mlu)

#kid_mom = zip(kid,mom)
#print(list(kid_mom))

    
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

numtics = len(kid)
x = np.arange(numtics)
plt.plot(x, kid)
plt.plot(x, mom)
plt.legend([child, child+'\'s mom'], loc='best')
plt.title('Mean Length of Utterance (MLU)')
plt.xlabel('Visit number')
plt.ylabel('MLU')

```
